```python
df['Position'] = df['Position'].str.replace('Centre-Back','CB')
df['Position'] = df['Position'].str.replace('Central Midfield','CM')
df['Position'] = df['Position'].str.replace('Centre-Forward','CF')
df['Position'] = df['Position'].str.replace('Right Winger','CB')
df['Position'] = df['Position'].str.replace('Left Winger','LW')
df['Position'] = df['Position'].str.replace('Attacking Midfield','AMF')
df['Position'] = df['Position'].str.replace('Defensive Midfield','DMF')
df['Position'] = df['Position'].str.replace('Right-Back','RB')
df['Position'] = df['Position'].str.replace('Left-Back','LB')
df['Position'] = df['Position'].str.replace('Goalkeeper','GK')
df['Position'] = df['Position'].str.replace('Second Striker','SS')
df['Position'] = df['Position'].str.replace('Left Midfield','LM')
df['Position'] = df['Position'].str.replace('Right Midfield','RM')
```

```python
df.head()
```

```python
df['Markey Value In Millions(£)'] = df['Markey Value In Millions(£)'].astype(int)
```

```python
df = df.rename(columns={'Markey Value In Millions(£)': 'Market Value'})
```

```python
df.head()
```

```python
df['Club'].value_counts()
```

```python
df['Goals per Game'] = df['Goals']/df['Matches']
df['Assists per Game'] = df['Assists']/df['Matches']
df
```

```python
df['Goals per Game'] = round(df['Goals per Game'],2)
df['Assists per Game']= round(df['Assists per Game'],2)
```

```python
df.head()
```

```python
df['Goal Contribution per Game'] = df['Goals per Game'] + df['Assists per Game']
```

```python
df.head()
```

```python
df['Playing full 90'] = df['Matches']-(df['Number Of Substitute Out']+df['Number Of Substitute In']) 
```

# **Visualization**

```python
import matplotlib.pyplot as plt
plt.hist(df['Market Value'], bins=20)
plt.xlabel('Market Value')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Position', y='Market Value')
plt.xlabel('Position')
plt.ylabel('Market Value')
plt.title('Market Value by Position')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Country', y='Market Value', ci=None)
plt.xlabel('Country')
plt.ylabel('Average Market Value')
plt.title('Market Value by Country')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.scatter(df['Age'], df['Market Value'])
plt.xlabel('Age')
plt.ylabel('Market Value')
plt.title('Market Value vs. Age')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Club', y='Market Value', ci=None)
plt.xlabel('Club')
plt.ylabel('Average Market Value')
plt.title('Market Value by Club')
plt.xticks(rotation=90)
plt.show()
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
import numpy as np

# Create a pivot table to summarize the average market value by position and age
pivot_table = df.pivot_table(values='Market Value', index='Position', columns='Age', aggfunc=np.mean)

# Create a heatmap to visualize the market value distribution
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt=".2f", cbar_kws={'label': 'Market Value'})
plt.title('Market Value Distribution by Position and Age')
plt.xlabel('Age')
plt.ylabel('Position')
plt.show()
```

```python
import plotly.express as px
fig = px.scatter(df, x='Goals', y='Market Value', title='Market Value vs. Goals',
                 labels={'Goals': 'Goals', 'Market Value': 'Market Value'})
fig.show()
```
